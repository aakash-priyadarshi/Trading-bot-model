```python
import pickle
import pandas as pd

# Load the trained model
with open('model.pkl', 'rb') as model_file:
    model = pickle.load(model_file)

# Function to make predictions
def predict(data):
    # Assuming 'data' is a DataFrame with the same structure as the training data
    predictions = model.predict(data)
    return predictions

# Example usage
if __name__ == "__main__":
    # Load your test data
    test_data = pd.read_csv('cleaned_new_stock_data_with_indicators.csv')
    
    # Make predictions
    predictions = predict(test_data)
    
    # Save predictions to a CSV file
    output = pd.DataFrame({'Date': test_data['Date'], 'Predicted_Close': predictions})
    output.to_csv('predictions.csv', index=False)
```
